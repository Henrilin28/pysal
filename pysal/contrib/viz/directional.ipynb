```python
import pysal as ps
import numpy as np
```

```python
f = ps.open(ps.examples.get_path("usjoin.csv"))
pci = np.array([f.by_col[str(y)] for y in range(1929,2010)])
```

```python
y0 = pci[0] # first year
y1 = pci[-1] #final year

# normalize to mean
y0 = y0 / y0.mean()
y1 = y1 / y1.mean()

```

```python
# get limits
minY = min ([ y0.min(), y1.min()])
maxY = max ([ y0.max(), y1.max()])
```

```python
w = ps.open(ps.examples.get_path("states48.gal")).read()
w.transform = 'r'

             
```

```python
wy0 = ps.lag_spatial(w, y0)
wy1 = ps.lag_spatial(w, y1)
```

```python
scatter(y0, wy0)
ylim([minY, maxY])
xlim([minY, maxY])
ylabel('wy0')
xlabel('y0')
```

```python
scatter(y1, wy1)
ylim([minY, maxY])
xlim([minY, maxY])
ylabel('wy1')
xlabel('y1')
```

```python
dy = y1 - y0
dwy = wy1 - wy0
# limits
mind = min([dy.min(), dwy.min()])
maxd = min([dy.max(), dwy.max()])
mabs = max(np.abs([mind, maxd]))
limits = [-mabs, mabs]
```

```python
mat = np.vstack([dy, dwy]).T
```
